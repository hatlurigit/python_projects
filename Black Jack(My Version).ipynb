Black Jack game:

Need:
    1. Deck
    2. value for each card
    3. adjust value for aces
    4. bet from the player
    5. dealing of cards
    6. hide the dealer card
    7. dealer will deal until the dealer cards hit 21

```python
import random

# numbers = ['Two','Three','Four','Five','Six','Seven','Eight','Nine']
# faces = ['Jack', 'Queen', 'King', 'Aces']
suites = ['Hearts', 'Diamonds', 'Spades', 'Clubs']
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 
            'Nine':9, 'Ten':10, 'Jack':10, 'Queen':10, 'King':10, 'Ace':[11,1]}

class Card:
    
    def __init__(self,number,suit):
        self.number = number
        self.suit = suit
        
        
    def __str__(self):
            return self.number +' of '+ self.suit
```

```python
class Deck:
    
    def __init__(self):
        deck = []
        for key in values.keys():
            for suite in suites:
                deck.append(Card(key,suite))
        self.deck = deck
        
        
    def __str__(self):
        deck_comp = ''
        for i in self.deck:
            deck_comp += "\n" + i.__str__()
            
        return "The deck has following cards" + deck_comp
```

```python
class Deal(Deck):
    
    def __init__(self):
        Deck.__init__(self)
        
    def shuffle(self):
        return random.shuffle(self.deck)
    
    def top_card(self):
        return self.deck.pop()
```

```python
class Hand:
    
    def __init__(self):
        self.cards = []
        self.total = 0
        
    def totals(self,card):
        self.cards.append(card)
        if card.number != 'Ace':
            self.total += values[card.number]
        elif card.number == 'Ace' and self.total < 21 and (self.total+10) <= 21:
            self.total += 10
        elif card.number == 'Ace' and self.total < 21 and (self.total+10) > 21:
            self.total += 1
        else:
            self.total += 10
```

```python
class Bet:
    
    def __init__(self):
        self.bet = 0
        self.total = 0
        self.percent_bet = 1.5
        
    def win_bet(self):
        self.total = self.bet + (self.bet* self.percent_bet)
        return self.total
        
    def loose_bet(self):
        self.total = self.bet - (self.bet *self.percent_bet)
        return self.total
```

Play Sequence:
    - take the bet
    - deal the cards
    - show the cards for both player and dealer
    - check totals
    - ask the player where he wants to continue getting cards or not
    - at the same time check if the total < 21
    - Once the player has stopped playing
    - Dealer will start with his cards:
        - show the second dealer cars
        - add the total
        - if the total > 21. stop the dealer cards.
        - if the total < player.total deal untill the total is greater than player or 18 whichever comes fast.
        - Once the dealer stops check whether the player has won
        - if player won give the bet money

```python
def bet_check():
    while True:
        try:
            bet = int(input("How much do you want to bet on this game:\n"))
        except ValueError:
            print("Sorry enter a valid number")
            continue
        else:
            print("You have placed the bet for: ", bet)
            break
    return int(bet)


def initial_deal(player_deck,dealer_deck,deck):
    for i in range(4):
        if i == 0:
            player_deck.totals(deck.top_card())
        elif i == 1:
            dealer_deck.totals(deck.top_card())
        elif i == 2:
            player_deck.totals(deck.top_card())
            print("\n\tPLAYER CARDS: ")
            for i in player_deck.cards:
                print(f"\n\t\t{i}")
            print(f"\n\tPLAYER CURRENT TOTAL: {player_deck.total}")
        else:
            dealer_deck.totals(deck.top_card())
            print("\n\tDEALER CARDS: ")
            for i in dealer_deck.cards:
                if i == dealer_deck.cards[0]:
                    print("\n\t\t<<<HIDDEN>>>")
                else:
                    print(f"\n\t\t{i}")
                    
def hit_stand(player_deck,deck):
    
        player_decision = True
        while player_deck.total < 21 and player_decision:
            player_turn = True
            while player_turn:
                player_decision = input("Please enter whether you want to Hit(H) or Stand(S): ")
                if player_decision == 'H':
                    player_deck.totals(deck.top_card())
                    print("\n\tPLAYER CARDS: ")
                    for i in player_deck.cards:
                        print(f"\n\t\t{i}")
                    print(f"\n\tPLAYER CURRENT TOTAL: {player_deck.total}")
                    break
                else:
                    print("\n\tPLAYER CARDS: ")
                    for i in player_deck.cards:
                        print(f"\n\t\t{i}")
                    print(f"\n\tPLAYER CURRENT TOTAL: {player_deck.total}\n\n<<<PLAYER TURN OVER.>>>")
                    player_turn = False
                    player_decision = False
        else:
            if player_deck.total > 21:
                print(f"\n\tPLAYER CURRENT TOTAL: {player_deck.total} \n\n<<<PLAYER BUSTED!!>>>")
                
                
def dealer_chance(dealer_deck,deck):
    
    print("\n\tDEALER CARDS: ")
    for i in dealer_deck.cards:
        print(f"\n\t\t{i}")
    print(f"\n\tDEALER CURRENT TOTAL: {dealer_deck.total}")
    dealer_turn = True
    while dealer_turn:
        if dealer_deck.total != 17 and dealer_deck.total <21 and dealer_deck.total < 17:
            dealer_deck.totals(deck.top_card())
            print("\n\tDEALER CARDS: ")
            for i in dealer_deck.cards:
                print(f"\n\t\t{i}")
            print(f"\n\tDEALER CURRENT TOTAL: {dealer_deck.total}")
        else:
            dealer_turn = False
        
        
        
```

```python
playing = True
while playing:
    player_bet = Bet()
    player_bet.bet = bet_check()
    playing_deck = Deal()
    playing_deck.shuffle()
    player_deck = Hand()
    dealer_deck = Hand()
    initial_deal(player_deck,dealer_deck,playing_deck)
    hit_stand(player_deck,playing_deck)

    if player_deck.total < 21:
        dealer_chance(dealer_deck,playing_deck)
        if dealer_deck.total > player_deck.total and dealer_deck.total <=21:
            print(f"\n\n<<<DEALER WINS>>>\nThe player looses: {player_bet.loose_bet()} dollars")
        else:
            print(f"\n\n<<<PLAYER WINS>>>\nThe player wins: {player_bet.win_bet()} dollars")         
    elif player_deck.total == 21:
        print(f"\n\n<<<PLAYER WINS>>>\nThe player wins: {player_bet.win_bet()} dollars")
    else:
        print(f"\n\n<<<DEALER WINS>>>\nThe player looses: {player_bet.loose_bet()} dollars")

    play_again = input("Do you want to play again?(Y/N): ")
    if play_again == 'Y':
        playing = True
    else:
        playing = False
```
